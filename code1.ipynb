```python
import jdatetime

def get_AD_date(date_str):
    year = int(date_str[:4])
    month = int(date_str[5:7])
    day = int(date_str[8:])
    jalali_date = jdatetime.date(year, month, day)
    gregorian_date = jalali_date.togregorian()
    return gregorian_date.strftime('%m/%d/%Y')
```

```python
mask = (dataframe == 0) | (pd.isnull(dataframe))
mask = ~mask
clean_df = dataframe.loc[mask.any(axis=1)]
clean_df = clean_df.reset_index(drop=True)
clean_df
```

دیتا تایپ های اعداد  یا فلوت هست یا اینت

```python
LEI_df = clean_df.copy()
```

```python
LEI_df.insert(4, 'First_AD_Date', '')

LEI_df.insert(6, 'Second_AD_Date', '')

LEI_df.insert(8, 'Third_AD_Date', '')

LEI_df.insert(10, 'first min cml thickness', '')
LEI_df.insert(11, 'second min cml thickness', '')
LEI_df.insert(12, 'third min cml thickness', '')
```

```python
LEI_df
```

```python
date_indexes = []
for i in range (len(LEI_df)):
    if len(str(LEI_df.at[i, 'G'])) == 10:
        date_indexes.append(i)
        # set the dates
        LEI_df.loc[i, 'First_AD_Date'] = get_AD_date(LEI_df.loc[i, 'G'])
        LEI_df.loc[i, 'Second_AD_Date'] = get_AD_date(LEI_df.loc[i, 'H'])
        LEI_df.loc[i, 'Third_AD_Date'] = get_AD_date(LEI_df.loc[i, 'I'])

        
```

بررسی درصد اندازه گیری های انجام شده

```python
for row in range(len(date_indexes)-1):
    start = date_indexes[row]
    stop = date_indexes[row+1]
    exl_file_name = LEI_df.loc[start-1, 'B']
    size = float(LEI_df.loc[start-1, 'B'].split('(')[1].split('-')[0])
    for i in range(start, stop-3, 4):
        if type(LEI_df.loc[i+1, 'Size (in)']) != str:
            if float(LEI_df.loc[i+1, 'Size (in)']) == size:
                LEI_df.loc[i+4, 'first min cml thickness'] = pd.to_numeric(LEI_df.loc[i+1:i+4, 'G'], errors='coerce').min()
                LEI_df.loc[i+4, 'second min cml thickness'] = pd.to_numeric(LEI_df.loc[i+1:i+4, 'H'], errors='coerce').min()
                LEI_df.loc[i+4, 'third min cml thickness'] = pd.to_numeric(LEI_df.loc[i+1:i+4, 'I'], errors='coerce').min()
        # ایجا اومدیم داخل 4 تا 4 تا هر قسمت
        measure1 = 0
        measure2 = 0
        measure3 = 0
        for j in range (i+1, i+5):
            if type(LEI_df.loc[j, 'G']) == float or type(LEI_df.loc[j, 'G']) == int:
                measure1 +=1
            LEI_df.loc[i+4, 'First_AD_Date'] = measure1/4

            if type(LEI_df.loc[j, 'H']) == float or type(LEI_df.loc[j, 'H']) == int:
                measure2 +=1
            LEI_df.loc[i+4, 'Second_AD_Date'] = measure2/4

            if type(LEI_df.loc[j, 'I']) == float or type(LEI_df.loc[j, 'I']) == int:
                measure3 +=1
            LEI_df.loc[i+4, 'Third_AD_Date'] = measure3/4

            
```

```python
export_df = pd.DataFrame(columns=["File name", "first inspection date", "second inspection date", "third inspection date", "first LEI", "second LEI", "third LEI", "first min thickness", "second min thickness", "third min thickness", "Max corr rate (mm/y)"])
```

```python
for row in range(len(date_indexes)-1):
    start = date_indexes[row]
    stop = date_indexes[row+1]
    export_df.loc[row, "File name"] = LEI_df.loc[start-1, 'B']
    export_df.loc[row, "first inspection date"] = LEI_df.loc[start, 'First_AD_Date']
    export_df.loc[row, "second inspection date"] = LEI_df.loc[start, 'Second_AD_Date']
    export_df.loc[row, "third inspection date"] = LEI_df.loc[start, 'Third_AD_Date']
    export_df.loc[row, "first min thickness"] = pd.to_numeric(LEI_df.loc[start+1:stop-2, 'first min cml thickness'], errors='coerce').min()
    export_df.loc[row, "second min thickness"] = pd.to_numeric(LEI_df.loc[start+1:stop-2, 'second min cml thickness'], errors='coerce').min()
    export_df.loc[row, "third min thickness"] = pd.to_numeric(LEI_df.loc[start+1:stop-2, 'third min cml thickness'], errors='coerce').min()
    export_df.loc[row, "Max corr rate (mm/y)"] = pd.to_numeric(LEI_df.loc[start+1:stop-2, 'Corr rate (mm/y)'], errors='coerce').max()
    #print(LEI_df.loc[start-1, 'B'])
    point_number = 0
    LEI1 = LEI2 = LEI3 = 0
    for i in range(start, stop-3, 4):
        var = LEI_df.loc[i+1, "B"]
        point_number += len(str(var).split(','))
        LEI1 += LEI_df.loc[i+4, "First_AD_Date"]*len(str(var).split(','))
        LEI2 += LEI_df.loc[i+4, "Second_AD_Date"]*len(str(var).split(','))
        LEI3 += LEI_df.loc[i+4, "Third_AD_Date"]*len(str(var).split(','))
    export_df.loc[row, "first LEI"] = LEI1/point_number
    export_df.loc[row, "second LEI"] = LEI2/point_number
    export_df.loc[row, "third LEI"] = LEI3/point_number
        
```

```python
for row in range(len(export_df)):
    file_name = export_df.loc[row, 'File name']
    for i in range (3):
        if i == 0:

            third_LEI = export_df.loc[row, "third LEI"]
            if third_LEI == 1:
                final_df.loc[(row*3)+i, "General Thinning"] = 'A - Highly Effective'
            elif third_LEI >= 0.75:
                final_df.loc[(row*3)+i, "General Thinning"] = 'B - Usually Effective'
            elif third_LEI >= 0.5:
                final_df.loc[(row*3)+i, "General Thinning"] = 'C - Fairly Effective'
            elif third_LEI >= 0.25:
                final_df.loc[(row*3)+i, "General Thinning"] = 'D - Poorly Effective'
            elif third_LEI < 0.25:
                final_df.loc[(row*3)+i, "General Thinning"] = 'E - Not Effective'

            final_df.loc[(row*3)+i, "Inspection Date"] = export_df.loc[row, "third inspection date"]
            
            final_df.loc[(row*3)+i, "train"] = file_name.split('(')[1].split(')')[0].split('-')[3][0]
            final_df.loc[(row*3)+i, "excell file name"] = file_name.split('(')[1].split(')')[0].split('-')[2]
            final_df.loc[(row*3)+i, "Equipment - Component Name"] = file_name.split('(')[1].split(')')[0].replace("-", "\"-", 1)
            final_df.loc[(row*3)+i, "Min Measured Thickness"] = export_df.loc[row, "third min thickness"]
            final_df.loc[(row*3)+i, "Comment"] = file_name
            final_df.loc[(row*3)+i, "maximum corrosion rate (mm/y)"] = export_df.loc[row, 'Max corr rate (mm/y)']
        elif i == 1:

            second_LEI = export_df.loc[row, "second LEI"]
            if second_LEI == 1:
                final_df.loc[(row*3)+i, "General Thinning"] = 'A - Highly Effective'
            elif second_LEI >= 0.75:
                final_df.loc[(row*3)+i, "General Thinning"] = 'B - Usually Effective'
            elif second_LEI >= 0.5:
                final_df.loc[(row*3)+i, "General Thinning"] = 'C - Fairly Effective'
            elif second_LEI >= 0.25:
                final_df.loc[(row*3)+i, "General Thinning"] = 'D - Poorly Effective'
            elif second_LEI < 0.25:
                final_df.loc[(row*3)+i, "General Thinning"] = 'E - Not Effective'
            final_df.loc[(row*3)+i, "train"] = file_name.split('(')[1].split(')')[0].split('-')[3][0]
            final_df.loc[(row*3)+i, "excell file name"] = file_name.split('(')[1].split(')')[0].split('-')[2]
            final_df.loc[(row*3)+i, "Equipment - Component Name"] = file_name.split('(')[1].split(')')[0].replace("-", "\"-", 1)
            final_df.loc[(row*3)+i, "Inspection Date"] = export_df.loc[row, "second inspection date"]
            final_df.loc[(row*3)+i, "Comment"] = file_name
            final_df.loc[(row*3)+i, "Min Measured Thickness"] = export_df.loc[row, "second min thickness"]
        elif i == 2:

            first_LEI = export_df.loc[row, "first LEI"]
            if first_LEI == 1:
                final_df.loc[(row*3)+i, "General Thinning"] = 'A - Highly Effective'
            elif first_LEI >= 0.75:
                final_df.loc[(row*3)+i, "General Thinning"] = 'B - Usually Effective'
            elif first_LEI >= 0.5:
                final_df.loc[(row*3)+i, "General Thinning"] = 'C - Fairly Effective'
            elif first_LEI >= 0.25:
                final_df.loc[(row*3)+i, "General Thinning"] = 'D - Poorly Effective'
            elif first_LEI < 0.25:
                final_df.loc[(row*3)+i, "General Thinning"] = 'E - Not Effective'
            final_df.loc[(row*3)+i, "Inspection Date"] = export_df.loc[row, "first inspection date"]
            final_df.loc[(row*3)+i, "train"] = file_name.split('(')[1].split(')')[0].split('-')[3][0]
            final_df.loc[(row*3)+i, "excell file name"] = file_name.split('(')[1].split(')')[0].split('-')[2]
            final_df.loc[(row*3)+i, "Equipment - Component Name"] = file_name.split('(')[1].split(')')[0].replace("-", "\"-", 1)
            final_df.loc[(row*3)+i, "Comment"] = file_name
            final_df.loc[(row*3)+i, "Min Measured Thickness"] = export_df.loc[row, "first min thickness"]

        
    
```
