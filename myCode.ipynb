```python
import pandas as pd
import os
```

```python
ExcellData = pd.read_excel('Sheet1.xlsx')
ExcellData
```

Making the new name for each file

```python
for i in range(len(ExcellData)):
    new_name =  f"{ExcellData.loc[i, "Line No"]} SHEET {ExcellData.loc[i, "Sheet"]} of {ExcellData.loc[i, "of"]}"
    ExcellData.loc[i, "previous name"] = f"{ExcellData.loc[i, "previous name"]}.pdf"
    # Exception for sheet (number) of A
    if new_name[-1] == "A":
        new_name = f"{ExcellData.loc[i, "Line No"]} SHEET {ExcellData.loc[i, "Sheet"]}A"
    
     
    ExcellData.loc[i, "new name"] = f"{new_name}.pdf"

    modified_name = ExcellData.loc[i, "new name"].replace('"', "''")
    modified_name = modified_name.replace('1/2', '0.5')
    modified_name = modified_name.replace('3/4', '0.75')

    ExcellData.loc[i, "modified name"] = modified_name
##############################################################################################
```

```python
ExcellData.to_excel('new names.xlsx')
```

```python
for i in range(len(ExcellData)):    
    previous_name = ExcellData.loc[i, "previous name"]
    new_name = ExcellData.loc[i, "modified name"]

    os.rename(previous_name, new_name)
```
