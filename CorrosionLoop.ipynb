```python
df = pd.read_excel("Book1.xlsx")
```

```python
for i in range (4122):
    target_string = str(df.loc[i, "line"])
    target_string = target_string.strip()
    filtered_df = df[df['Line number'] == target_string]

    if not filtered_df.empty:
        row_location = filtered_df.index[0]
        df.loc[i, "Loop name"] = df.loc[row_location, "Name of Loop"]
    else:
        df.loc[i, "Loop name"] = ("not found in the corrosion loop.")
```
